```python
# trying to merge the data from the different months
import requests
import pandas as pd
from datetime import datetime, timedelta
import os

# Function to generate date ranges by month
def month_range(start_date, end_date):
    current_month = start_date.replace(day=1)
    while current_month < end_date:
        next_month = current_month.replace(day=28) + timedelta(days=4)
        yield current_month, min(next_month - timedelta(days=1), end_date)
        current_month = next_month

# API endpoint
url = "http://api.worldweatheronline.com/premium/v1/past-weather.ashx"

# load API key from text file
with open('api_key_weather.txt', 'r') as file:
    api_key = file.read().replace('\n', '')

# Create a folder to store monthly data
if not os.path.exists('Months'):
    os.makedirs('Months')

# Define the timeframe for data fetching
start_date = datetime(2021, 5, 1)
end_date = datetime(2022, 8, 31)

# Loop through each month in the timeframe
for start_month, end_month in month_range(start_date, end_date):
    # Parameters
    params = {
        "key": str(api_key),
        "q": "Dublin",  # Query location
        "format": "json",
        "date": start_month.strftime("%Y-%m-%d"),  # Start date for historical data
        "enddate": end_month.strftime("%Y-%m-%d"),  # End date for historical data
        "includelocation": "yes",
        "tp": "1"  # Time period: 1 hour
    }

    # Making the GET request
    response = requests.get(url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()

        # Extracting location data
        location_data = data.get('data', {}).get('nearest_area', [{}])[0]
        latitude = location_data.get('latitude', 'Unknown')
        longitude = location_data.get('longitude', 'Unknown')

        # Extracting the weather data
        weather_data = data.get('data', {}).get('weather', [])

        # Creating a list to store each day's weather information
        weather_list = []
        for day in weather_data:
            hourly_list = day.get('hourly', [])
            for hourly_data in hourly_list:
                hour_value = hourly_data.get('time', 'Unknown')  # This should extract the time for each hour
                if hour_value != 'Unknown':
                    hour_formatted = f"{int(hour_value)//100:02d}:00" if hour_value != '0' else "00:00"
                else:
                    hour_formatted = hour_value
                day_data = {
                    'date': day['date'],
                    'hour': hour_formatted,
                    'avgtempC': day['avgtempC'],
                    'maxtempC': day['maxtempC'],
                    'mintempC': day['mintempC'],
                    'sunHour': day['sunHour'],
                    'uvIndex': day['uvIndex'],
                    'humidity': hourly_data['humidity'],
                    'winddirDegree': hourly_data['winddirDegree'],
                    'windspeedKmph': hourly_data['windspeedKmph'],
                    'cloudcover': hourly_data['cloudcover'],
                    'precipMM': hourly_data['precipMM'],
                    'pressure': hourly_data['pressure'],
                    'latitude': latitude,
                    'longitude': longitude,
                }
                weather_list.append(day_data)

        # Combine date and hour into a datetime object
        for item in weather_list:
            date_str = item['date']  # 'YYYY-MM-DD'
            hour_str = item['hour']  # 'HH:00' format
            datetime_str = f"{date_str} {hour_str}"
            item['datetime'] = datetime.strptime(datetime_str, "%Y-%m-%d %H:%M")

        # Convert the list of dictionaries to a pandas DataFrame
        weather_df = pd.DataFrame(weather_list)

        # Set 'datetime' as the index of the DataFrame
        weather_df.set_index('datetime', inplace=True)

        # Save the DataFrame to a CSV file
        folder_name = 'Months'
        file_name = f'dublin_weather_{start_month.strftime("%Y-%m")}.csv'
        file_path = os.path.join(folder_name, file_name)
        weather_df.to_csv(file_path, index=True)

    else:
        print(f"Failed to fetch data for {start_month.strftime('%Y-%m')}. Status Code:", response.status_code)
```

```python
# Folder containing monthly CSV files
folder_name = 'Months'

# List to store DataFrames from each CSV file
dfs = []

# Loop through each CSV file in the folder
for file_name in os.listdir(folder_name):
    if file_name.endswith('.csv'):
        # Read the CSV file into a DataFrame
        file_path = os.path.join(folder_name, file_name)
        df = pd.read_csv(file_path, parse_dates=['datetime'], index_col='datetime')
        dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dfs)

# Sort the DataFrame by the datetime index
merged_df.sort_index(inplace=True)

# Print the first few rows of the merged DataFrame
print(merged_df.head())

# Optionally, save the merged DataFrame to a CSV file
merged_file_path = 'merged_weather_data.csv'
merged_df.to_csv(merged_file_path)
```

```python
# checking that every hour of every day are present in the dataset 
import pandas as pd
from datetime import datetime, timedelta

# Function to generate date range
def date_range(start_date, end_date):
    for n in range(int((end_date - start_date).days) + 1):
        yield start_date + timedelta(n)

# Load the merged DataFrame
merged_file_path = 'merged_weather_data.csv'
merged_df = pd.read_csv(merged_file_path, parse_dates=['datetime'], index_col='datetime')

# Define start and end dates
start_date = datetime(2021, 5, 1)
end_date = datetime(2022, 8, 31)

# Check if every day and every hour between start_date and end_date is present in the DataFrame
missing_datetimes = []
for single_date in date_range(start_date, end_date):
    for hour in range(24):
        current_datetime = single_date.replace(hour=hour)
        if current_datetime not in merged_df.index:
            missing_datetimes.append(current_datetime)

# Print the missing datetimes
if missing_datetimes:
    print("Missing datetimes:")
    for dt in missing_datetimes:
        print(dt)
else:
    print("All datetimes are present in the DataFrame.")
```
