```python
#libraries needed for scraping and handling data
import requests
import pandas as pd
```

```python
#API access Sensor.Community
#https://github.com/opendata-stuttgart/meta/wiki/APIs

#this API key gives access to average of all measurements per sensor of the 24 hours
api_24h_mean = 'https://data.sensor.community/static/v2/data.1h.json'

#this API key gives access to average of all measurements per sensor of the last hour.
api_1h_mean = 'https://data.sensor.community/static/v2/data.1h.json'
```

```python
#response (json) is stored in variable
response24 = requests.get(api_24h_mean).json()
response1 = requests.get(api_1h_mean).json()
```

```python
#this snippet iterates through the json response, unpacks specified data and saves it in lists

id_list = []
sampling_rate_list = []
timestamp_list = []
location_list = []
sensor_list = []
sensordatavalues_list = []
location_ids = []
latitude = []
longitude = []
altitude = []
country = []
exact_location = []
indoor = []
sensor_ids = []
pin = []
sensor_type_id = []
sensor_type_name = []
sensor_type_manufacturer = []
humiditys = []
temperatures = []
pressures = []
SDS_P1s = []
SDS_P2s = []
P0s = []
P1s = []
P2s = []
P3s = []
P4s = []
N1s = []
N4s = []
N10s = []
N25s = []


for item in response24:
    iD = item['id']
    id_list.append(iD)
    sampling_rate = item['sampling_rate']
    sampling_rate_list.append(sampling_rate)
    timestamp = item['timestamp']
    timestamp_list.append(timestamp)
    location = item['location']
    for key, value in location.items():
        if key == 'id':
            location_ids.append(value)
        elif key == 'latitude':
            latitude.append(value)
        elif key == 'longitude':
            longitude.append(value)
        elif key == 'altitude':
            altitude.append(value)
        elif key == 'country':
            country.append(value)
        elif key == 'exact_location':
            exact_location.append(value)
        elif key == 'indoor':
            indoor.append(value)
    location_list.append(location)
    sensor = item['sensor']
    for key, value in sensor.items():
        if key == 'id':
            sensor_ids.append(value)
        elif key == 'pin':
            pin.append(value)
        elif key == 'sensor_type':
            for sub_key, sub_value in value.items():
                if sub_key == 'id':
                    sensor_type_id.append(sub_value)
                elif sub_key == 'name':
                    sensor_type_name.append(sub_value)
                elif sub_key == 'manufacturer':
                    sensor_type_manufacturer.append(sub_value)
    sensor_list.append(sensor)
    sensordatavalues = item['sensordatavalues']
    value_types = ['humidity','temperature','pressure','P0','P1','P2','P3','P4','N1','N4','N10','N25']
    result = {key: 'NA' for key in value_types}
    for data in sensordatavalues:
        if data['value_type'] in result:
            result[data['value_type']] = data['value']
    humidity = [result['humidity']][0]
    humiditys.append(humidity)
    temperature = [result['temperature']][0]
    temperatures.append(temperature)
    pressure = [result['pressure']][0]
    pressures.append(pressure)
    P0 = [result['P0']][0]
    P0s.append(P0)
    P1 = [result['P1']][0]
    P1s.append(P1)
    P2 = [result['P2']][0]
    P2s.append(P2)
    P3 = [result['P3']][0]
    P3s.append(P3)
    P4 = [result['P4']][0]
    P4s.append(P4)
    N1 = [result['N1']][0]
    N1s.append(N1)
    N4 = [result['N4']][0]
    N4s.append(N4)
    N10 = [result['N10']][0]
    N10s.append(N10)
    N25 = [result['N25']][0]
    N25s.append(N25)
    sensordatavalues_list.append(sensordatavalues)
```

```python
#datacollection is formated as pandas dataframe
df = pd.DataFrame(list(zip(id_list, sensor_ids,sampling_rate_list,timestamp_list, sensor_type_name, 
                               location_list,sensor_list,sensordatavalues_list,latitude, longitude, country,
                               humiditys, temperatures, pressures,P0s,P1s,P2s,P3s,P4s,N1s,N4s,N10s,N25s)),
               columns =['id','sensor_id','sampling_rate','timestamp','sensor_type_name','location','sensor','sensordatavalues','latitude', 'longitude', 
                         'country','humidity','temperature','pressure','P0','PM10','PM2.5','P3','P4','N1','N4','N10','N25'])
```

```python
# import module
#!pip install geopy

#this function automatically assigns a city, town or county based on lat,long values
#caution: super slow! only one request per second
from geopy.geocoders import Nominatim

def assign_city(df):
    geolocator = Nominatim(user_agent="geoapiExercises")
    
    city_list = []
    for index,row in df.iterrows():
        lat = row['latitude']
        long = row['longitude']
        location = geolocator.reverse(f"{lat},{long}", exactly_one = True)
        city = location.raw['address'].get('town',"") or location.raw['address'].get('city', "") or location.raw['address'].get('county', "")
        city_list.append(city)
        
    df['city'] = city_list
    return df
```
